```python
df = X.loc[:,['price','horsepower','engine-size','curb-weight']]
```

```python
df[:2]
```

```python
df.dropna(inplace=True)
```

```python
x = df.drop('price',axis= 1)
y = df['price']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
ml_model = LinearRegression()
```

```python
ml_model
```

```python
ml_model.fit(x,y)
```

```python
ml_model.coef_
```

```python
ml_model.intercept_
```

```python
df[:2]
```

```python
ml_model.predict([[111,130,2548]])[0]
```

```python
ml_model.score(x,y)*100
```
